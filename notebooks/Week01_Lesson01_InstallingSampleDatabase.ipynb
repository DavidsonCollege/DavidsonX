# Week 1 - Lesson xx - Installing the DVD Rentals sample database

## Unzipping the dataset

First, go to the folder where you downloaded the course files from GitHub, find the "dvdrental" ZIP file, and unzip its contents into a new folder named "dvdrental."

<br>

The dvdrental folder should contain a single file, <code>dvdrental.tar</code>, which is a backup file that contains the database definition and all the data needed to create our sample database. Make note of the folder path for this file, or copy it to your clipboard, because you will need it in a moment in order to create our database using this backup file.

## Create the database in postgres

Execute the SQL below to connect to the "postgres" database and create the "dvdrental" database.

````sql
-- this will create a new database named "dvdrental"
CREATE DATABASE dvdrental;
````

Once the database has been created, open your terminal by going to the View menu or by using Ctrl+`, then paste the following command and run it by hitting Enter.

Make sure to replace the {path to repository} with the folder location where you unzipped

```
pg_restore --username=postgres --dbname=dvdrental --verbose "C:\{path to repository}\DavidsonX-SQL-for-Everyone\data\dvdrental\dvdrental.tar" 
```

You will be prompted for the password you gave when you installed Postgres on your computer. Type that password into the terminal when prompted to do so, then hit Enter.

At this point, your terminal should fill up with dozens of lines of output text, saying <code>pg_restore: creating [something]</code>. This means the restore has worked successfully. If you go to your Connections panel in Azure Data Studio and refresh the "Databases" folder under your Postgres database, you should now see a new database named "dvdrental."
